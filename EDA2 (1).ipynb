```python
import numpy as np
adult = adult_income.copy()

# Feature creation
adult['capital_net'] = adult['capital_gain'] - adult['capital_loss']
adult['age_hours_ratio'] = adult['hours_per_week'] / adult['age']

# Log transformations
adult['log_capital_gain'] = np.log1p(adult['capital_gain'])
adult['log_capital_loss'] = np.log1p(adult['capital_loss'])
```

```python
adult
```

```python
from sklearn.ensemble import IsolationForest

# Select numeric features
num_cols = ['age', 'education_num', 'capital_gain', 'capital_loss', 'hours_per_week']
X = adult_income[num_cols]

# Fit Isolation Forest
iso = IsolationForest(n_estimators=100, contamination=0.01, random_state=42)
adult_income['outlier'] = iso.fit_predict(X)

# Remove flagged outliers (-1)
clean_df = adult_income[adult_income['outlier'] == 1].copy()
print(f"Removed {sum(adult_income['outlier'] == -1)} outliers")
```
