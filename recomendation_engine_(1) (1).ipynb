```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# Genre to TF-IDF
tfidf = TfidfVectorizer()
genre_vec = tfidf.fit_transform(anime['genre']).toarray()

# Normalize rating
scaler = MinMaxScaler()
anime['rating_norm'] = scaler.fit_transform(anime[['rating']])
anime['rating_norm'].fillna(0, inplace=True) # Ensure NaNs are filled

# Combine
features = np.hstack([genre_vec, anime['rating_norm'].values.reshape(-1,1)])
```

```python
cos_sim = cosine_similarity(features, features)

def recommend(anime_name, k=2):
    idx = anime[anime['name']==anime_name].index[0]
    sim_scores = list(enumerate(cos_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    top = [anime.iloc[i]['name'] for i,_ in sim_scores[1:k+1]]
    return top
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score

# Suppose we had ground truth of relevant animes (manually or from user data)
true_relevant = {"Fullmetal Alchemist", "Code Geass", "Naruto"}
recommended = set(recommend("Fullmetal Alchemist: Brotherhood", k=10))
```

```python
# Convert to binary for evaluation
y_true = [1 if anime in true_relevant else 0 for anime in recommended]
y_pred = [1] * len(y_true)  # Since all are predicted
```

```python
# Evaluate
print("\nSimulated Evaluation:")
print(f"Precision: {precision_score(y_true, y_pred):.2f}")
print(f"Recall: {recall_score(y_true, y_pred):.2f}")
print(f"F1-score : {f1_score(y_true, y_pred):.2f}")
```

# **Q1 :**

Collaborative filtering splits into two popular approaches:

1 User-Based Collaborative Filtering (UBCF)

Concept: Find users similar to the target user, then recommend items those users liked.

Example: If User A and User B both like Movies X and Y, and User B also likes Movie Z, suggest Z to User A.

Similarity: Computed via Pearson correlation or cosine similarity on user rating vectors

2 Item-Based Collaborative Filtering (IBCF) :

Concept: Look for items similar to those the target user has interacted with, regardless of user identity.

Example: If many users who bought Item X also bought Item Y, recommend Y to anyone who buys X.

Collaborative Filtering (CF) is a technique used in recommender systems to predict a user's interests by leveraging the preferences or behaviors of many users. It operates under the principle that users with similar tastes are likely to enjoy similar items .

collaborative filtering shines at delivering novel content — even items without explicit features, by relying purely on user behavior data
